### DETERMINE THE TOP THREE MOST COMMON CUISINES IN THE DATASET.


```python
df['Cuisines'].value_counts()
```

```python
df['Cuisines'].value_counts().head(3)
```

```python
Top3_Cuisines = df['Cuisines'].value_counts().head(3)
```

```python
plt.figure(figsize=(10,6))
Top3_Cuisines.plot(kind ='bar', color ='skyblue', edgecolor ='black')
plt.title("Top 3 Cuisines of the Restaurants")
plt.xlabel('Cuisines')
plt.ylabel("Count")
plt.xticks(rotation =45)    #  For Better Readability
plt.show()
```
